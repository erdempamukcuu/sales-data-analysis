```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data load
df = pd.read_csv("data.csv")
```

```python
# First Look
print(df.head())
```

```python
# Total sales
total_sales = df['Sales_Amount'].sum()
print("Total Sales_Amount:", total_sales)
```

```python
# Product-based sales
product_sales = df.groupby('Product_Category')['Sales_Amount'].sum()
product_sales
```

```python
product_sales.plot(kind='bar')
plt.title("Sales per Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
